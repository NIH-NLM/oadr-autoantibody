```python
import numpy as np

def aggregate_weights_median(weights_dict):
    """
    Aggregate weights layer-wise across all studies using the median.
    """
    studies = list(weights_dict.keys())
    n_layers = len([w for w in weights_dict[studies[0]] if w is not None])

    aggregated_weights = []
    for layer_idx in range(n_layers):
        # Collect all weights for this layer across studies
        layer_weights = [weights_dict[study][layer_idx] for study in studies if weights_dict[study] is not None]

        # Stack and take median across the first axis (studies)
        stacked = np.stack(layer_weights, axis=0)
        median_weight = np.median(stacked, axis=0)
        aggregated_weights.append(median_weight)

    return aggregated_weights
```

```python
def extract_model_weights_per_study(mse_summaries):
    weights_dict = {}

    for study in mse_summaries.keys():
        model_path = f"models/{study}_model.keras"

        try:
            model = load_model(model_path)
            weights = model.get_weights()
            weights_dict[study] = weights
        except Exception as e:
            print(f"Could not load model for {study}: {e}")
            weights_dict[study] = None

    return weights_dict
```

```python
from tensorflow.keras.models import load_model
import os

def extract_model_weights_per_study(study_list, model_dir="models"):
    """
    Extract weights from saved models for each study in study_list.

    Parameters:
        study_list (list of str): List of study IDs (e.g., ["SDY569", ...])
        model_dir (str): Directory where model files are saved.

    Returns:
        dict: Mapping of study ID to its model weights.
    """
    weights_dict = {}

    for study in study_list:
        model_path = os.path.join(model_dir, f"{study}_3x3_model.keras")

        try:
            model = load_model(model_path)
            weights = model.get_weights()
            weights_dict[study] = weights
        except Exception as e:
            print(f"⚠️ Could not load model for {study}: {e}")
            weights_dict[study] = None

    return weights_dict
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.models import model_from_json
import os

def build_model_from_file(model_path):
    """
    Load a Keras model structure from an existing model file (without weights).
    """
    model = load_model(model_path)
    model_json = model.to_json()
    new_model = model_from_json(model_json)
    return new_model
```

```python
def build_federated_model(aggregated_weights, reference_model_path):
    """
    Build a model with aggregated weights.
    """
    model = build_model_from_file(reference_model_path)
    model.set_weights(aggregated_weights)
    return model
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

def evaluate_federated_model_3x3(federated_model, study_id):
    # Load test data
    df_test = pd.read_csv(f"test_data/{study_id}_3x3_test.csv")

    # Extract features
    X_test = df_test.iloc[:, :9].values  # full input
    X_test_reshaped = X_test.reshape((-1, 3, 3, 1))

    # Predict
    reconstructed = federated_model.predict(X_test_reshaped)
    reconstructed_flat = reconstructed.reshape(X_test.shape)

    # Evaluate first 5 columns (autoantibodies)
    ab_cols = ["GAD65", "IA2IC", "MIAA", "ICA", "ZNT8"]
    X_ab_true = X_test[:, :5]
    X_ab_pred = reconstructed_flat[:, :5]

    # Overall MSE
    mse = mean_squared_error(X_ab_true, X_ab_pred)

    # Per-antibody MSE
    per_antibody_mse = {
        ab: mean_squared_error(X_ab_true[:, i], X_ab_pred[:, i])
        for i, ab in enumerate(ab_cols)
    }
    mse_ab_median = np.median(list(per_antibody_mse.values()))
    mse_ab_iqr = np.percentile(list(per_antibody_mse.values()), 75) - \
                 np.percentile(list(per_antibody_mse.values()), 25)

    # Reconstruct DataFrames
    df_true = pd.DataFrame(X_ab_true, columns=ab_cols)
    df_pred = pd.DataFrame(X_ab_pred, columns=ab_cols)

    # Reconstruct Age_Group from one-hot encoded columns
    age_group_cols = ["8-12", "13-17", "18-30"]
    df_test["Age_Group"] = df_test[age_group_cols].idxmax(axis=1)
    
    # Reconstruct Sex from binary column
    df_test["Sex"] = df_test["Sex"].map({0: "Male", 1: "Female"})

    # Attach original demographic values
    df_true["Age_Group"] = df_test["Age_Group"]
    df_true["Sex"] = df_test["Sex"]
    df_pred["Age_Group"] = df_test["Age_Group"]
    df_pred["Sex"] = df_test["Sex"]

    # MSE by age group
    mse_by_age_group = df_true.groupby("Age_Group").apply(
        lambda g: mean_squared_error(
            g[ab_cols], df_pred.loc[g.index, ab_cols]
        )
    ).to_dict()

    mse_age_group_median = np.median(list(mse_by_age_group.values()))
    mse_age_group_iqr = np.percentile(list(mse_by_age_group.values()), 75) - \
                        np.percentile(list(mse_by_age_group.values()), 25)

    # MSE by sex
    mse_by_sex = df_true.groupby("Sex").apply(
        lambda g: mean_squared_error(
            g[ab_cols], df_pred.loc[g.index, ab_cols]
        )
    ).to_dict()

    mse_sex_median = np.median(list(mse_by_sex.values()))
    mse_sex_iqr = np.percentile(list(mse_by_sex.values()), 75) - \
                  np.percentile(list(mse_by_sex.values()), 25)

    return {
        "study": study_id,
        "mse": mse,
        "mse_ab_median": mse_ab_median,
        "mse_ab_iqr": mse_ab_iqr,
        "per_antibody_mse": per_antibody_mse,
        "mse_by_age_group": mse_by_age_group,
        "mse_by_sex": mse_by_sex,
        "mse_age_group_median": mse_age_group_median,
        "mse_age_group_iqr": mse_age_group_iqr,
        "mse_sex_median": mse_sex_median,
        "mse_sex_iqr": mse_sex_iqr,
        "N_test": len(X_test),
        "y_true": X_ab_true,
        "y_pred": X_ab_pred,
    }
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

def augment_local_mse_summary(summary):
    ab_cols = summary["antibody_labels"]
    X_test = summary["X_test"]
    X_pred = summary["reconstructed"][:, :5]  # only antibody part
    age_groups = summary["Age_Group"]
    sex = summary["Sex"]

    # Rebuild dataframes
    df_true = pd.DataFrame(X_test, columns=ab_cols)
    df_pred = pd.DataFrame(X_pred, columns=ab_cols)
    df_true["Age_Group"] = age_groups
    df_pred["Age_Group"] = age_groups
    df_true["Sex"] = sex
    df_pred["Sex"] = sex

    # MSE by age group
    mse_by_age_group = df_true.groupby("Age_Group").apply(
        lambda g: mean_squared_error(
            g[ab_cols], df_pred.loc[g.index, ab_cols]
        )
    ).to_dict()

    mse_age_group_median = np.median(list(mse_by_age_group.values()))
    mse_age_group_iqr = np.percentile(list(mse_by_age_group.values()), 75) - \
                        np.percentile(list(mse_by_age_group.values()), 25)

    # MSE by sex
    mse_by_sex = df_true.groupby("Sex").apply(
        lambda g: mean_squared_error(
            g[ab_cols], df_pred.loc[g.index, ab_cols]
        )
    ).to_dict()

    mse_sex_median = np.median(list(mse_by_sex.values()))
    mse_sex_iqr = np.percentile(list(mse_by_sex.values()), 75) - \
                  np.percentile(list(mse_by_sex.values()), 25)

    # Update the summary dict
    summary.update({
        "mse_by_age_group": mse_by_age_group,
        "mse_age_group_median": mse_age_group_median,
        "mse_age_group_iqr": mse_age_group_iqr,
        "mse_by_sex": mse_by_sex,
        "mse_sex_median": mse_sex_median,
        "mse_sex_iqr": mse_sex_iqr,
    })

    return summary
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_comparison_medians(mse_summaries_local, mse_summaries_fed, study_list):
    metrics = [
        ("mse_ab_median", "Median MSE (Antibodies)"),
        ("mse_age_group_median", "Median MSE by Age Group"),
        ("mse_sex_median", "Median MSE by Sex")
    ]
    iqrs = [
        ("mse_ab_iqr", "Antibody IQR"),
        ("mse_age_group_iqr", "Age Group IQR"),
        ("mse_sex_iqr", "Sex IQR")
    ]

    # Map local key equivalents
    local_key_map = {
        "mse_ab_median": "mse_median",
        "mse_ab_iqr": "mse_iqr",
        "mse_age_group_median": "mse_age_group_median",
        "mse_age_group_iqr": "mse_age_group_iqr",
        "mse_sex_median": "mse_sex_median",
        "mse_sex_iqr": "mse_sex_iqr"
    }

    x = np.arange(len(study_list))
    width = 0.35

    for (metric_key, ylabel), (iqr_key, _) in zip(metrics, iqrs):
        local_metric_key = local_key_map.get(metric_key, metric_key)
        local_iqr_key = local_key_map.get(iqr_key, iqr_key)

        try:
            local_vals = [mse_summaries_local[study][local_metric_key] for study in study_list]
            fed_vals = [mse_summaries_fed[study][metric_key] for study in study_list]
            local_iqrs = [mse_summaries_local[study][local_iqr_key] for study in study_list]
            fed_iqrs = [mse_summaries_fed[study][iqr_key] for study in study_list]
        except KeyError as e:
            print(f"Missing key in summaries: {e}")
            continue

        fig, ax = plt.subplots(figsize=(10, 6))
        bars1 = ax.bar(x - width/2, local_vals, width, yerr=local_iqrs, label='Local', capsize=5)
        bars2 = ax.bar(x + width/2, fed_vals, width, yerr=fed_iqrs, label='Federated', capsize=5)

        ax.set_ylabel(ylabel)
        ax.set_title(f"{ylabel} – Local vs Federated")
        ax.set_xticks(x)
        ax.set_xticklabels(study_list, rotation=45)
        ax.legend()
        ax.grid(axis='y', linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()
```

```python
def flatten_mse_results(results_dict, model_type="local"):
    records = []

    for study, res in results_dict.items():
        row = {
            "Study": study,
            "Model_Type": model_type,
            "N_test": res.get("N_test", len(res.get("X_test", []))),  # fallback
            "MSE_All": res.get("mse"),
            "MSE_AB_Median": res.get("mse_ab_median", res.get("mse_median")),
            "MSE_AB_IQR": res.get("mse_ab_iqr", res.get("mse_iqr")),
            "MSE_AgeGroup_Median": res.get("mse_age_group_median"),
            "MSE_AgeGroup_IQR": res.get("mse_age_group_iqr"),
            "MSE_Sex_Median": res.get("mse_sex_median"),
            "MSE_Sex_IQR": res.get("mse_sex_iqr"),
        }

        for ab, ab_mse in res.get("per_antibody_mse", {}).items():
            row[f"MSE_{ab}"] = ab_mse

        records.append(row)

    return pd.DataFrame(records)
```

```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from tensorflow.keras import layers, models


def train_and_evaluate_cnn_3x3(df, study_list, study_name, epochs=50, initial_weights=None):
    print(f"\n===== Training 3x3 CNN Autoencoder for {study_name} =====")

    if isinstance(study_list, str):
        study_list = [study_list]

    # --- Filter study ---
    df_filtered = df[df["Study"].isin(study_list)].copy()
    if df_filtered.empty:
        raise ValueError(f"No data found for study {study_name}")

    # --- Normalize antibody names ---
    df_filtered["Property"] = (
        df_filtered["Property"]
        .str.upper()
        .str.replace(r"[^A-Z0-9]", "", regex=True)
    )

    # --- Pivot to wide format ---
    df_wide = df_filtered.pivot_table(
        index="Accession",
        columns="Property",
        values="Value",
        aggfunc="first"
    ).fillna(0)

    # --- Expected antibody columns ---
    antibody_labels = ["GAD65", "IA2IC", "MIAA", "ICA", "ZNT8"]
    df_wide = df_wide.reindex(columns=antibody_labels, fill_value=0)

    # --- Join demographics ---
    demo_cols = (
        df_filtered[["Accession", "Age_Group", "Sex"]]
        .drop_duplicates()
        .set_index("Accession")
    )
    df_wide = df_wide.join(demo_cols, how="left")

    # --- Scale antibody values ---
    ab_scaler = MinMaxScaler()
    X_ab = ab_scaler.fit_transform(df_wide[antibody_labels].values)

    # --- One-hot encode demographics ---
    demo_array = df_wide[["Age_Group", "Sex"]].fillna("missing").astype(str).values
    encoder = OneHotEncoder(sparse_output=False)
    X_demo = encoder.fit_transform(demo_array)

    # Pad or trim to ensure 4 demographic features
    if X_demo.shape[1] < 4:
        pad_width = 4 - X_demo.shape[1]
        X_demo = np.pad(X_demo, ((0, 0), (0, pad_width)), mode="constant")
    elif X_demo.shape[1] > 4:
        X_demo = X_demo[:, :4]

    # --- Combine into final input ---
    X_full = np.hstack([X_ab, X_demo]).astype(np.float32)
    if X_full.shape[1] != 9:
        raise ValueError(f"Expected 9 features, got {X_full.shape[1]}")

    # --- Train/test split ---
    indices = np.arange(X_full.shape[0])
    X_train, X_test, idx_train, idx_test = train_test_split(
        X_full, indices, test_size=0.5, random_state=42
    )

    # --- Reshape to 3×3 image ---
    X_train_img = X_train.reshape(-1, 3, 3, 1)
    X_test_img = X_test.reshape(-1, 3, 3, 1)

    # --- CNN Autoencoder ---
    inputs = layers.Input(shape=(3, 3, 1))
    x = layers.Conv2D(8, (3, 3), activation="relu", padding="same")(inputs)
    x = layers.Flatten()(x)
    encoded = layers.Dense(4, activation="relu")(x)
    x = layers.Dense(9, activation="relu")(encoded)
    x = layers.Reshape((3, 3, 1))(x)
    outputs = layers.Conv2D(1, (3, 3), activation="linear", padding="same")(x)
    autoencoder = models.Model(inputs, outputs)
    autoencoder.compile(optimizer="adam", loss="mse")

    if initial_weights is not None:
        autoencoder.set_weights(initial_weights)

    # --- Train ---
    autoencoder.fit(
        X_train_img, X_train_img,
        epochs=epochs,
        batch_size=8,
        validation_data=(X_test_img, X_test_img),
        verbose=0
    )

    # --- Predict ---
    reconstructed = autoencoder.predict(X_test_img)
    reconstructed_flat = reconstructed.reshape(X_test.shape)

    # --- MSE calculations ---
    per_antibody_mse = {
        ab: mean_squared_error(X_test[:, i], reconstructed_flat[:, i])
        for i, ab in enumerate(antibody_labels)
    }
    mse_ab_median = np.median(list(per_antibody_mse.values()))
    mse_ab_iqr = np.percentile(list(per_antibody_mse.values()), 75) - \
                 np.percentile(list(per_antibody_mse.values()), 25)

    # --- Age/Sex MSE analysis ---
    
    df_test = df_wide.iloc[idx_test]
    age_groups = df_test["Age_Group"].values
    sex = df_test["Sex"].values

    df_true = pd.DataFrame(X_test[:, :5], columns=antibody_labels)
    df_pred = pd.DataFrame(reconstructed_flat[:, :5], columns=antibody_labels)
    df_true["Age_Group"] = age_groups
    df_true["Sex"] = sex
    df_pred["Age_Group"] = age_groups
    df_pred["Sex"] = sex

    # MSE by age
    mse_by_age_group = df_true.groupby("Age_Group").apply(
        lambda g: mean_squared_error(g[antibody_labels], df_pred.loc[g.index, antibody_labels])
    ).to_dict()
    mse_age_group_median = np.median(list(mse_by_age_group.values()))
    mse_age_group_iqr = np.percentile(list(mse_by_age_group.values()), 75) - \
                        np.percentile(list(mse_by_age_group.values()), 25)

    # MSE by sex
    mse_by_sex = df_true.groupby("Sex").apply(
        lambda g: mean_squared_error(g[antibody_labels], df_pred.loc[g.index, antibody_labels])
    ).to_dict()
    mse_sex_median = np.median(list(mse_by_sex.values()))
    mse_sex_iqr = np.percentile(list(mse_by_sex.values()), 75) - \
                  np.percentile(list(mse_by_sex.values()), 25)

    # --- Save model ---
    model_dir = "models"
    os.makedirs(model_dir, exist_ok=True)
    model_path = os.path.join(model_dir, f"{study_name}_3x3_model.keras")
    weights_path = os.path.join(model_dir, f"{study_name}_3x3.weights.h5")
    autoencoder.save(model_path)
    autoencoder.save_weights(weights_path)
    print(f"Saved model:    {model_path}")
    print(f"Saved weights:  {weights_path}")
    print(f"MSE_AB_Median: {mse_ab_median:.6f} | IQR: {mse_ab_iqr:.6f}")

    # --- Save test data for reuse ---
    output_dir = "test_data"
    os.makedirs(output_dir, exist_ok=True)
    demo_feature_names = ["8-12", "13-17", "18-30", "Sex"]
    input_feature_names = antibody_labels + demo_feature_names
    df_test_out = pd.DataFrame(X_test, columns=input_feature_names)
    df_test_out["Accession"] = df_wide.index[idx_test]  # add it explicitly

    # --- Optional: Sanity check for alignment ---
    if True:  # Change to False to disable
        df_check = pd.DataFrame(X_test, columns=input_feature_names)
        df_check["Accession"] = df_wide.index[idx_test]
        print("Sample aligned test features:")
        display(df_check.head())
        print("Sample source data for those accessions:")
        display(df_wide.loc[df_check["Accession"]].head())
    
    csv_path = os.path.join(output_dir, f"{study_name}_3x3_test.csv")
    df_test_out.to_csv(csv_path, index=False)
    print(f"Saved test data CSV: {csv_path}")

    return {
        "mse_ab_median": mse_ab_median,
        "mse_ab_iqr": mse_ab_iqr,
        "per_antibody_mse": per_antibody_mse,
        "mse_by_age_group": mse_by_age_group,
        "mse_by_sex": mse_by_sex,
        "mse_age_group_median": mse_age_group_median,
        "mse_age_group_iqr": mse_age_group_iqr,
        "mse_sex_median": mse_sex_median,
        "mse_sex_iqr": mse_sex_iqr,
        "X_train": X_train,
        "X_test": X_test,
        "reconstructed": reconstructed_flat,
        "antibody_labels": antibody_labels,
        "Age_Group": age_groups,
        "Sex": sex
    }
```

```python
# lets look at each of the antibodies individually
def flatten_per_antibody_mse(all_results_dict, model_type):
    """
    Converts per-antibody MSEs into a long-form DataFrame for comparison.

    Parameters:
        all_results_dict (dict): Dictionary with structure {study_id: result_dict}
        model_type (str): "local", "federated", or "retrained"

    Returns:
        pd.DataFrame with columns: Study, Model_Type, Antibody, MSE
    """
    records = []

    for study, res in all_results_dict.items():
        per_ab_mse = res.get("per_antibody_mse", {})
        for ab, mse in per_ab_mse.items():
            records.append({
                "Study": study,
                "Model_Type": model_type,
                "Antibody": ab,
                "MSE": mse
            })

    return pd.DataFrame(records)
```

## Predicting residual beta-cell function from autoantibody profiles, under federation

### Why this prediction matters

In Type 1 Diabetes the body's own immune system progressively destroys the
insulin-producing cells of the pancreas (the beta cells of the islets of
Langerhans). The disease's natural history is the slow disappearance of those
cells; in long-standing Type 1 Diabetes there are essentially none left.

The most informative quantitative readout of how many functional beta cells a
patient still has is the **C-peptide** they release during a standardised
four-hour meal challenge.  C-peptide is the short piece that gets cleaved off
when proinsulin matures into insulin, so it is released into the bloodstream in
exact one-to-one proportion with insulin.  Unlike insulin, it is not removed
by the liver in the first pass and not confounded by injected insulin
treatment, which makes the area under its four-hour blood-concentration curve
the field-standard measurement of how much insulin the patient's own pancreas
is still capable of making.  This area-under-the-curve number is what we
predict here, on a log scale so the distribution is roughly normal.

The five autoantibodies in the input panel (GAD65, IA-2, Insulin/MIAA, ICA and
ZnT8) are the immune-system signatures of the same destructive process: their
levels rise during the attack that destroys the beta cells.  Whether and how
well an autoantibody profile predicts the beta-cell function that survives
that attack is what every model below asks.

### Why federation

Each patient cohort in this analysis comes from a different institutional
clinical study (SDY524, SDY569, SDY797, SDY1737).  These
data sets cannot be moved between institutions: privacy regulations, IRB
constraints, and institutional policies forbid it.  

**federated learning**

1. **No subject data leaves its home institution.**  Each study trains its own
   local model on its own subjects only.
2. **Only parameter vectors are exchanged.**  After local training each
   institution sends its model's parameters (a vector of numbers) to a
   coordinator.
3. **Aggregation is by per-parameter median across institutions.**  For each
   parameter the coordinator takes the middle value across studies; the median
   is robust to one institution being unusual (for example SDY797's
   autoantibody columns are coded 0/1 instead of continuous).
4. **At prediction time each subject is rescaled by the local scaler of its
   own institution** before the federated parameters are applied.  Scalers
   never leave their study either.

### What this section measures

Two arguments are tested numerically:

> *(a) Federating with more institutions produces better predictions, even
> when each institution only trains once.*

To test this we sweep the number of participating studies from one to four.
At each count we average the prediction error over every choice of study
combination of that size, so the number is not biased by which subset we
picked.

> *(b) Iterating the federation across multiple rounds extracts further
> benefit until the improvement curve flattens.*

To test this we run the convolutional methods for several rounds: round one
trains from scratch; subsequent rounds let each institution warm-start its
local model from the previous round's aggregated parameters and continue
training, then re-federate.  We stop when the round-over-round change in
prediction error falls below a small threshold.  The simple-linear and lasso
methods are not iterated, since their per-study fits are closed-form
solutions that do not benefit from continuation.

Four prediction methods are compared:

- **Simple linear regression.** The prediction is a weighted sum of the input
  features plus an intercept: y = b0 + b1*x1 + b2*x2 + ... + bk*xk. Each
  institution fits its own weights; the coordinator takes the per-weight
  median across institutions.
- **Lasso regression.** "Lasso" stands for *Least Absolute Shrinkage
  and Selection Operator*. Same equation as simple linear regression, but the
  fitting procedure penalises large weights and pushes weak features to
  exactly zero.  The non-zero weights after federation are the features that
  survived selection in a majority of institutions.
- **Convolutional neural network.** A small image-style model that treats the
  nine-element feature vector as a 3-by-3 grid, applies a small convolution,
  and learns a non-linear mapping to C-peptide directly.  It is the
  "naive deep learning" baseline.
- **Random Forest (used here for *regression*, not classification).**
  A Random Forest is an ensemble of decision trees: each tree splits
  the feature space recursively to predict the target. Random Forests
  are most commonly seen in classification tasks, and even then a
  formal p-value requires permuting the labels and re-fitting the
  forest hundreds or thousands of times. Here we are not extracting
  a p-value: the forest is used as a *prediction model* for the
  continuous C-peptide target. The forest still produces a
  feature-importance ranking (the fraction of variance reduction each
  feature contributes), which is what makes it interpretable. Federation
  is done by taking every institution's forest of trees and averaging
  predictions across all of them; Random Forest is invariant to feature
  scaling so no scaler is needed.
- **Convolutional neural network with autoencoder pretraining.** Same
  architecture, but training is split into two stages.  Stage one trains
  *only* the feature-encoding half of the network to faithfully reconstruct
  the inputs, without ever seeing C-peptide.  Stage two freezes that encoder
  and trains a small predictive head on top against C-peptide.  The
  autoencoder gives the predictive head a stable, low-dimensional summary
  of the antibody profile instead of forcing the head to learn that summary
  jointly with the prediction task on tiny per-institution sample sizes.

### Effective sample size

Panel A in the shared loader is the four-study cohort with one hundred and
fifty subjects total.  Per-institution sample sizes are very small (SDY569
has ten subjects with C-peptide; SDY1737 has sixteen), which is precisely
the regime where federation is supposed to help.

### Model definitions in plain language

Below, each method is built as a small function that returns a ready-to-train
model.  All four methods predict the same target (log C-peptide) from the
same nine-feature input; they differ only in how the prediction is
constructed.

- The two **convolutional** networks reshape the nine numeric features into
  a 3-by-3 grid and apply a single small convolution before flattening
  back to a vector.  Reshaping like this is the simplest way to let a
  convolutional layer see the input; nothing physical about the grid is
  meaningful.
- The **autoencoder** is just a convolutional network whose output has the
  same shape as the input and which is trained to reproduce the input.
  The middle (smallest) layer is a four-number summary called the
  *latent representation*.  After training, we discard the output half and
  freeze the encoder half: it becomes a frozen feature extractor that
  turns the nine input numbers into four summary numbers.  A small
  predictive head sits on top.


```python
# === Model factories ===

def build_3x3_autoencoder():
    # Convolutional autoencoder.  Trained only to reconstruct its input
    #  (no labels).  Architecture: Conv2D(8 filters) -> Flatten -> Dense(4)
    #   -> Dense(9) -> Reshape -> Conv2D(1 filter).
    #  The Dense(4) layer is the four-number latent summary; everything from
    #  the input up to that layer is the encoder.
    inp = layers.Input(shape=(3, 3, 1))
    x = layers.Conv2D(8, (3, 3), activation='relu', padding='same', name='enc_conv')(inp)
    x = layers.Flatten(name='enc_flat')(x)
    z = layers.Dense(4, activation='relu', name='enc_latent')(x)
    x = layers.Dense(9, activation='relu', name='dec_dense')(z)
    x = layers.Reshape((3, 3, 1), name='dec_reshape')(x)
    out = layers.Conv2D(1, (3, 3), activation='linear', padding='same', name='dec_conv')(x)
    ae = models.Model(inp, out, name='autoencoder')
    ae.compile(optimizer='adam', loss='mse')
    return ae


def build_head_from_autoencoder_weights(ae_weights):
    # Build a predictive head atop a frozen federated encoder.
    # The encoder layers are loaded from the supplied (federated) autoencoder
    # weights; the head is a fresh Dense(16) -> Dense(1) stack with 97
    #  trainable parameters
    ae = build_3x3_autoencoder()
    ae.set_weights(ae_weights)
    inp = layers.Input(shape=(3, 3, 1))
    x = ae.get_layer('enc_conv')(inp)
    x = ae.get_layer('enc_flat')(x)
    z = ae.get_layer('enc_latent')(x)
    h = layers.Dense(16, activation='relu', name='head_hidden')(z)
    out = layers.Dense(1, name='head_out')(h)
    m = models.Model(inp, out, name='ae_head')
    for n in ('enc_conv', 'enc_latent'):
        m.get_layer(n).trainable = False
    m.compile(optimizer='adam', loss='mse')
    return m


def build_plain_cnn():
    # End-to-end convolutional network with no autoencoder pretraining.
    # Conv2D(16 filters) -> Flatten -> Dense(32) -> Dense(1).
    # Trained directly against the C-peptide target.
    m = models.Sequential([
        layers.Input(shape=(3, 3, 1)),
        layers.Conv2D(16, (3, 3), activation='relu', padding='same'),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),
        layers.Dense(1),
    ])
    m.compile(optimizer='adam', loss='mse')
    return m


def aggregate_by_median(parameters_by_study):
    # Federated aggregation: take the element-wise median across institutions
    # for each parameter tensor in the model.
    studies = list(parameters_by_study.keys())
    n_layers = len(parameters_by_study[studies[0]])
    out = []
    for layer_idx in range(n_layers):
        stacked = np.stack([parameters_by_study[s][layer_idx] for s in studies], axis=0)
        out.append(np.median(stacked, axis=0))
    return out
```

### Per-method federation procedure

Each method below follows the same outline:

1. For each institution: scale features locally, fit the model locally.
2. Collect parameters; aggregate by per-parameter median.
3. At prediction time: for each test subject, look up the subject's home
   institution, apply that institution's local scaler, then evaluate the
   federated model on the scaled vector.

The simple-linear and lasso methods only need one round (a closed-form fit).
The two convolutional methods can be warm-started for additional rounds:
each institution receives the previous round's federated parameters,
continues training locally, and sends parameters back.


```python
# === Federated fit/predict, one per method ===

def fit_federated_simple_linear(X, y, study_labels):
    """ Each institution fits ordinary linear regression on its own subjects;
        the coordinator takes the per-coefficient median across institutions. """
    coefs, scalers = {}, {}
    for s in sorted(np.unique(study_labels)):
        mask = (study_labels == s)
        if mask.sum() < 2:
            continue
        sc = MinMaxScaler().fit(X[mask])
        lr = LinearRegression().fit(sc.transform(X[mask]), y[mask])
        scalers[s] = sc
        coefs[s] = np.concatenate(([lr.intercept_], lr.coef_))
    fed = np.median(np.stack(list(coefs.values()), axis=0), axis=0)
    return {'type': 'linear', 'intercept': fed[0], 'beta': fed[1:], 'scalers': scalers}


def fit_federated_lasso(X, y, study_labels, alpha=LASSO_ALPHA):
    """Each institution fits a lasso regression at a shared regularisation
       strength; the coordinator takes the per-coefficient median.  Coefficients
       that are pushed to zero in a majority of institutions stay zero after
       aggregation."""
    coefs, scalers = {}, {}
    for s in sorted(np.unique(study_labels)):
        mask = (study_labels == s)
        if mask.sum() < 2:
            continue
        sc = MinMaxScaler().fit(X[mask])
        lasso = Lasso(alpha=alpha, max_iter=20000, random_state=RNG_SEED).fit(sc.transform(X[mask]), y[mask])
        scalers[s] = sc
        coefs[s] = np.concatenate(([lasso.intercept_], lasso.coef_))
    fed = np.median(np.stack(list(coefs.values()), axis=0), axis=0)
    return {'type': 'lasso', 'intercept': fed[0], 'beta': fed[1:], 'scalers': scalers}




def fit_federated_random_forest(X, y, study_labels, n_trees=RF_N_TREES, seed=RNG_SEED):
    """Each institution trains its own Random Forest of n_trees decision trees
    on its own subjects. The federated 'model' is the collection of every
    institution's forest. Random Forest is invariant to feature scale so no
    MinMaxScaler is used."""
    forests = {}
    for s in sorted(np.unique(study_labels)):
        mask = (study_labels == s)
        if mask.sum() < 3:
            continue
        rf = RandomForestRegressor(n_estimators=n_trees, min_samples_leaf=2,
                                   random_state=seed, n_jobs=1).fit(X[mask], y[mask])
        forests[s] = rf
    return {'type': 'random_forest', 'forests': forests}


def fit_federated_plain_cnn(X, y, study_labels,
                             init_weights=None, epochs=CNN_EPOCHS_COLD,
                             fold_idx=0, seed=RNG_SEED):
    """Each institution trains the convolutional network end-to-end on its
    own subjects.  When init_weights is supplied (rounds >= 2) the model
    starts from the previous round's federated parameters."""
    weights, scalers = {}, {}
    for s_idx, s in enumerate(sorted(np.unique(study_labels))):
        mask = (study_labels == s)
        if mask.sum() < 2:
            continue
        sc = MinMaxScaler().fit(X[mask])
        X_s = sc.transform(X[mask]).reshape(-1, 3, 3, 1)
        tf.keras.utils.set_random_seed(seed + fold_idx * 100 + s_idx)
        cnn = build_plain_cnn()
        if init_weights is not None:
            cnn.set_weights(init_weights)
        cnn.fit(X_s, y[mask], epochs=epochs, batch_size=BATCH, verbose=0)
        scalers[s] = sc
        weights[s] = cnn.get_weights()
    return {'type': 'plain_cnn',
            'weights': aggregate_by_median(weights),
            'scalers': scalers}


def fit_federated_autoencoder_cnn(X, y, study_labels,
                                   init_ae_weights=None, init_head_weights=None,
                                   ae_epochs=AE_EPOCHS_COLD, head_epochs=HEAD_EPOCHS_COLD,
                                   fold_idx=0, seed=RNG_SEED):
    """Two-stage federated training.

    Stage one: each institution trains its own autoencoder to reconstruct
    its own inputs (no C-peptide labels used).  The coordinator takes the
    per-weight median across institutions; this becomes the federated
    autoencoder.

    Stage two: each institution builds a predictive head on top of the
    *federated* (frozen) encoder, trains the head on its own (X, y)
    subjects, and sends head parameters back.  The coordinator again takes
    the per-weight median.  Because every institution's head sees the same
    encoder representation, median aggregation of the head parameters
    is well posed."""
    
    scalers, local_ae, local_head = {}, {}, {}
    # Stage 1: per-institution autoencoder
    for s_idx, s in enumerate(sorted(np.unique(study_labels))):
        mask = (study_labels == s)
        if mask.sum() < 2:
            continue
        sc = MinMaxScaler().fit(X[mask])
        X_s = sc.transform(X[mask]).reshape(-1, 3, 3, 1)
        tf.keras.utils.set_random_seed(seed + fold_idx * 100 + s_idx)
        ae = build_3x3_autoencoder()
        if init_ae_weights is not None:
            ae.set_weights(init_ae_weights)
        ae.fit(X_s, X_s, epochs=ae_epochs, batch_size=BATCH, verbose=0)
        scalers[s] = sc
        local_ae[s] = ae.get_weights()
    fed_ae = aggregate_by_median(local_ae)
    # Stage 2: per-institution predictive head atop the federated encoder
    for s_idx, s in enumerate(sorted(np.unique(study_labels))):
        mask = (study_labels == s)
        if mask.sum() < 2:
            continue
        X_s = scalers[s].transform(X[mask]).reshape(-1, 3, 3, 1)
        tf.keras.utils.set_random_seed(seed + fold_idx * 100 + s_idx + 500)
        head = build_head_from_autoencoder_weights(fed_ae)
        if init_head_weights is not None:
            head.set_weights(init_head_weights)
        head.fit(X_s, y[mask], epochs=head_epochs, batch_size=BATCH, verbose=0)
        local_head[s] = head.get_weights()
    return {'type': 'autoencoder_cnn',
            'ae_weights': fed_ae,
            'head_weights': aggregate_by_median(local_head),
            'scalers': scalers}


def predict_federated(state, X, study_labels):
    """Apply a federated model to test subjects.  Each subject is scaled by
    its home institution's local scaler before the federated model
    evaluates it."""
    yhat = np.empty(len(X))
    if state['type'] in ('linear', 'lasso'):
        for s in np.unique(study_labels):
            idx = np.where(study_labels == s)[0]
            sc = state['scalers'].get(s)
            X_s = sc.transform(X[idx]) if sc is not None else X[idx]
            yhat[idx] = state['intercept'] + X_s.dot(state['beta'])
        return yhat
    if state['type'] == 'random_forest':
        preds = np.zeros(len(X))
        for rf in state['forests'].values():
            preds += rf.predict(X)
        return preds / len(state['forests'])
    if state['type'] == 'plain_cnn':
        cnn = build_plain_cnn(); cnn.set_weights(state['weights'])
        for s in np.unique(study_labels):
            idx = np.where(study_labels == s)[0]
            sc = state['scalers'].get(s)
            X_s = (sc.transform(X[idx]) if sc is not None else X[idx]).reshape(-1, 3, 3, 1)
            yhat[idx] = cnn.predict(X_s, verbose=0).ravel()
        return yhat
    if state['type'] == 'autoencoder_cnn':
        head = build_head_from_autoencoder_weights(state['ae_weights'])
        head.set_weights(state['head_weights'])
        for s in np.unique(study_labels):
            idx = np.where(study_labels == s)[0]
            sc = state['scalers'].get(s)
            X_s = (sc.transform(X[idx]) if sc is not None else X[idx]).reshape(-1, 3, 3, 1)
            yhat[idx] = head.predict(X_s, verbose=0).ravel()
        return yhat
    raise ValueError(state['type'])
```

### Experiment

Two sweeps, both running five-fold cross-validation per setting:

- The **k-axis** asks how the mean prediction error changes as the
  number of participating institutions grows from one to four.  At each
  count we enumerate every combination of that size from the four
  institutions and average the prediction error.  (At one institution
  there are four subsets to average over; at two institutions there are
  six; at three there are four; at four there is one.)
- The **rounds-axis** asks how the prediction error of the two
  convolutional methods improves when each institution warm-starts from
  the previous round's aggregated parameters and re-trains.  We stop the
  iteration the first time the round-over-round change in mean
  prediction error falls below five thousandths of a unit.


```python
# === Experiment runners ===

def make_cv_iterator(X, y, study_labels, seed=RNG_SEED, n_splits=N_FOLDS):
    """Five-fold cross-validation, stratified by study when there is more
    than one institution in the subset (so every fold contains subjects
    from every institution in proportion); plain K-fold otherwise."""
    if len(np.unique(study_labels)) >= 2:
        return StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed).split(X, study_labels)
    return KFold(n_splits=n_splits, shuffle=True, random_state=seed).split(X)


def cv_one_subset_single_round(method_name, X, y, study_labels, subset_studies, seed=RNG_SEED):
    """Run five-fold cross-validation on the subjects belonging to the
    given subset of institutions.  Single round, no warm-start."""
    mask = np.isin(study_labels, subset_studies)
    Xs, ys, ss = X[mask], y[mask], study_labels[mask]
    fold_mses = []
    for fold_idx, (tr, te) in enumerate(make_cv_iterator(Xs, ys, ss, seed)):
        if method_name == 'simple_linear':
            state = fit_federated_simple_linear(Xs[tr], ys[tr], ss[tr])
        elif method_name == 'lasso':
            state = fit_federated_lasso(Xs[tr], ys[tr], ss[tr])
        elif method_name == 'random_forest':
            state = fit_federated_random_forest(Xs[tr], ys[tr], ss[tr])
        elif method_name == 'plain_cnn':
            state = fit_federated_plain_cnn(Xs[tr], ys[tr], ss[tr], fold_idx=fold_idx, seed=seed)
        elif method_name == 'autoencoder_cnn':
            state = fit_federated_autoencoder_cnn(Xs[tr], ys[tr], ss[tr], fold_idx=fold_idx, seed=seed)
        else:
            raise ValueError(method_name)
        yhat = predict_federated(state, Xs[te], ss[te])
        fold_mses.append(mean_squared_error(ys[te], yhat))
    return float(np.mean(fold_mses)), fold_mses


def run_k_axis_sweep(method_name, X, y, study_labels, k_values=(1, 2, 3, 4)):
    """For each k in (1, 2, 3, 4): enumerate every k-subset of institutions
    and average the cross-validation prediction error across subsets."""
    rows = []
    for k in k_values:
        subsets = list(combinations(ALL_STUDIES, k))
        subset_mses = []
        for sub in subsets:
            mean_mse, _ = cv_one_subset_single_round(method_name, X, y, study_labels, sub)
            subset_mses.append(mean_mse)
        rows.append({
            'method': method_name, 'k': k, 'n_subsets': len(subsets),
            'mean_MSE_over_subsets': float(np.mean(subset_mses)),
            'std_MSE_over_subsets': float(np.std(subset_mses)),
            'min_MSE_over_subsets': float(np.min(subset_mses)),
            'max_MSE_over_subsets': float(np.max(subset_mses)),
            'subset_MSEs': json.dumps(subset_mses),
        })
    return pd.DataFrame(rows)


def run_rounds_sweep_plain_cnn(X, y, study_labels, max_rounds=MAX_ROUNDS, plateau=PLATEAU_THRESHOLD):
    rows, prev_mean = [], None
    fed_w_prev = None
    for r in range(1, max_rounds + 1):
        cv_iter = list(make_cv_iterator(X, y, study_labels))
        fold_mses, per_fold = [], []
        for fold_idx, (tr, te) in enumerate(cv_iter):
            init = fed_w_prev[fold_idx] if fed_w_prev else None
            eps = CNN_EPOCHS_COLD if r == 1 else CNN_EPOCHS_WARM
            state = fit_federated_plain_cnn(X[tr], y[tr], study_labels[tr],
                                             init_weights=init, epochs=eps, fold_idx=fold_idx)
            yhat = predict_federated(state, X[te], study_labels[te])
            fold_mses.append(mean_squared_error(y[te], yhat))
            per_fold.append(state['weights'])
        mean = float(np.mean(fold_mses))
        rows.append({'method': 'plain_cnn', 'round': r, 'mean_MSE': mean,
                     'std_MSE': float(np.std(fold_mses)),
                     'fold_MSEs': json.dumps(fold_mses)})
        fed_w_prev = per_fold
        if prev_mean is not None and abs(prev_mean - mean) < plateau:
            break
        prev_mean = mean
    return pd.DataFrame(rows)


def run_rounds_sweep_autoencoder_cnn(X, y, study_labels, max_rounds=MAX_ROUNDS, plateau=PLATEAU_THRESHOLD):
    rows, prev_mean = [], None
    fed_ae_prev, fed_head_prev = None, None
    for r in range(1, max_rounds + 1):
        cv_iter = list(make_cv_iterator(X, y, study_labels))
        fold_mses, per_fold_ae, per_fold_head = [], [], []
        for fold_idx, (tr, te) in enumerate(cv_iter):
            init_ae = fed_ae_prev[fold_idx] if fed_ae_prev else None
            init_head = fed_head_prev[fold_idx] if fed_head_prev else None
            ae_eps = AE_EPOCHS_COLD if r == 1 else AE_EPOCHS_WARM
            head_eps = HEAD_EPOCHS_COLD if r == 1 else HEAD_EPOCHS_WARM
            state = fit_federated_autoencoder_cnn(X[tr], y[tr], study_labels[tr],
                                                   init_ae_weights=init_ae,
                                                   init_head_weights=init_head,
                                                   ae_epochs=ae_eps, head_epochs=head_eps,
                                                   fold_idx=fold_idx)
            yhat = predict_federated(state, X[te], study_labels[te])
            fold_mses.append(mean_squared_error(y[te], yhat))
            per_fold_ae.append(state['ae_weights'])
            per_fold_head.append(state['head_weights'])
        mean = float(np.mean(fold_mses))
        rows.append({'method': 'autoencoder_cnn', 'round': r, 'mean_MSE': mean,
                     'std_MSE': float(np.std(fold_mses)),
                     'fold_MSEs': json.dumps(fold_mses)})
        fed_ae_prev, fed_head_prev = per_fold_ae, per_fold_head
        if prev_mean is not None and abs(prev_mean - mean) < plateau:
            break
        prev_mean = mean
    return pd.DataFrame(rows)
```

### Which institution is added first matters

Averaging the prediction error over every subset of size k (as the
figure above does) gives the cleanest summary, but it hides which
specific institution is being added at each step. The two paths
below show the ordering explicitly: one starts with the largest
institution by sample size and adds smaller ones in turn; the other
starts with the smallest and adds larger ones.

Reading these paths together answers the question *"does my small
institution benefit from joining a federation with larger partners?"*
which is the operational question for the institutions that will
actually participate in the deployment.


```python
# === Figure: ordered federation paths ===
# Two orderings of the four institutions:
#   - 'Largest first': SDY524 (75) -> SDY797 (49) -> SDY1737 (16) -> SDY569 (10)
#   - 'Smallest first': SDY569 (10) -> SDY1737 (16) -> SDY797 (49) -> SDY524 (75)
# At each step we read the cached MSE for the exact subset from the k-axis CSV,
# so no re-training is needed.

study_n = {'SDY524': 75, 'SDY797': 49, 'SDY1737': 16, 'SDY569': 10}
order_largest_first  = ['SDY524', 'SDY797', 'SDY1737', 'SDY569']
order_smallest_first = ['SDY569', 'SDY1737', 'SDY797', 'SDY524']

def mse_for_specific_subset(method_name, subset):
    """Look up the cached MSE for one specific subset of institutions."""
    row = k_axis_df[(k_axis_df['method'] == method_name) & (k_axis_df['k'] == len(subset))].iloc[0]
    labels = json.loads(row['subset_labels'])
    mses = json.loads(row['subset_MSEs'])
    for i, sub in enumerate(labels):
        if set(sub) == set(subset):
            return mses[i]
    raise KeyError(subset)

all_methods_ordered = ['simple_linear', 'lasso', 'random_forest', 'plain_cnn', 'autoencoder_cnn']

fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
for ax, order, title in [
    (axes[0], order_largest_first,  'Starting with the largest institution, adding smaller ones'),
    (axes[1], order_smallest_first, 'Starting with the smallest institution, adding larger ones'),
]:
    for m in all_methods_ordered:
        path_subsets = [order[:i] for i in range(1, 5)]
        mses = [mse_for_specific_subset(m, sub) for sub in path_subsets]
        ax.plot(range(1, 5), mses, marker='o', color=method_color[m],
                label=method_display[m], linewidth=2.2, markersize=10)
    xlabels = []
    for i in range(4):
        cum = sum(study_n[s] for s in order[:i+1])
        if i == 0:
            xlabels.append(f'{order[i]}\n(N={cum})')
        else:
            xlabels.append(f'+{order[i]}\n(N={cum})')
    ax.set_xticks(range(1, 5))
    ax.set_xticklabels(xlabels, fontsize=9)
    ax.set_xlabel('Federation composition (cumulative subject count)')
    ax.set_title(title, fontsize=11)
    ax.grid(True, alpha=0.3)
axes[0].set_ylabel('Mean prediction error on log C-peptide')
axes[0].legend(loc='upper right', frameon=True, fontsize=9)
fig.suptitle('Which institution is added first matters: ordered federation paths', fontsize=13, y=1.01)
fig.tight_layout()
fig.savefig('figures/federated_ordered_paths.pdf', dpi=300, bbox_inches='tight')
fig.savefig('figures/federated_ordered_paths.png', dpi=220, bbox_inches='tight')
plt.show()
print('Saved figures/federated_ordered_paths.{pdf,png}')
```

### Federation buys statistical power

The two figures above measure the prediction error of various models
under federation. There is a parallel and equally fundamental argument
to be made about *statistical power*: at a fixed biological effect
size, the probability of detecting that effect with a hypothesis test
rises rapidly as the cohort size grows.

Multiple-regression power for a model with k features at sample size N
and effect size Cohen's f² is governed by the non-central F
distribution. The non-centrality parameter equals f² × N, so doubling
N at fixed f² substantially increases the power.

Federation is exactly the device that lets us grow N without any
single institution acquiring more subjects. The cell below computes
the power at the four cumulative cohort sizes produced by the
smallest-first ordered path (10, 26, 75, 150 subjects), for a range
of biologically plausible effect sizes.


```python
# === Federation buys statistical power ===
from scipy.stats import f as fdist, ncf

ALPHA = 0.05
K_FEATURES = len(fa)   # 9 features in Panel A

def calc_power(n, k, f2, alpha=ALPHA):
    """Power of an F-test for multiple regression with k predictors at
    sample size n and Cohen's f² effect size. Non-centrality nc = f²·N.
    """
    if n <= k + 1 or f2 <= 0:
        return float('nan')
    F_crit = fdist.ppf(1 - alpha, k, n - k - 1)
    nc = f2 * n
    return float(1 - ncf.cdf(F_crit, k, n - k - 1, nc))

def f2_for_r2(r2):
    return r2 / (1 - r2)

def mde_f2(n, k, target_power=0.80, alpha=ALPHA):
    """Minimum detectable Cohen's f² at the requested power, bisected."""
    lo, hi = 1e-6, 10.0
    if calc_power(n, k, hi, alpha) < target_power:
        return float('nan')
    for _ in range(60):
        mid = 0.5 * (lo + hi)
        if calc_power(n, k, mid, alpha) >= target_power:
            hi = mid
        else:
            lo = mid
    return 0.5 * (lo + hi)

# Cumulative cohort sizes along the smallest-first federation path.
cohort_sizes = [10, 26, 75, 150]
path_labels  = ['SDY569 alone', '+SDY1737', '+SDY797', '+SDY524 (full)']

# Tabulate power at typical biological effect sizes.
r2_values = [0.05, 0.10, 0.15, 0.20, 0.30]
rows = []
for r2 in r2_values:
    f2 = f2_for_r2(r2)
    for n, label in zip(cohort_sizes, path_labels):
        rows.append({'R2': r2, 'Cohen_f2': round(f2, 4), 'N': n,
                     'federation_state': label,
                     'power': calc_power(n, K_FEATURES, f2)})
power_df = pd.DataFrame(rows)
power_df.to_csv('results/federated_power_by_cohort_size.csv', index=False)
print('Power at federation cohort sizes (k = 9 features, alpha = 0.05):')
print(power_df.pivot(index=['R2', 'Cohen_f2'], columns='N', values='power').round(3).to_string())

print()
print('Minimum detectable effect size at 80% power, by federation state:')
for n, label in zip(cohort_sizes, path_labels):
    mde = mde_f2(n, K_FEATURES)
    if np.isnan(mde):
        print(f"  N={n:3d} ({label:18s}): cannot reach 80% power -- too few subjects to fit 9 features")
    else:
        print(f"  N={n:3d} ({label:18s}): f² >= {mde:.3f}  (equivalent R² >= {mde / (1 + mde):.3f})")
```

```python
# === Figure: power as a function of cohort size, at several effect sizes ===
fig, ax = plt.subplots(figsize=(10, 6))
n_grid = np.arange(K_FEATURES + 2, 200, 1)
for r2 in r2_values:
    f2 = f2_for_r2(r2)
    powers = [calc_power(n, K_FEATURES, f2) for n in n_grid]
    ax.plot(n_grid, powers, linewidth=2,
            label=f'effect size R² = {r2:.2f}  (Cohen f² = {f2:.3f})')
for n, label in zip(cohort_sizes, path_labels):
    ax.axvline(x=n, color='gray', linestyle='--', alpha=0.4)
    ax.text(n, 1.02, label, rotation=45, ha='left', va='bottom',
            fontsize=8.5, color='dimgray')
ax.axhline(y=0.80, color='black', linestyle=':', alpha=0.5, label='80% power threshold')
ax.set_xlabel('Cohort size (total subjects across all participating institutions)')
ax.set_ylabel('Statistical power to detect an effect of the stated size')
ax.set_title('Federation buys statistical power: more institutions -> higher power at every effect size\n(multiple regression with 9 features, α = 0.05)')
ax.set_xlim(10, 200)
ax.set_ylim(0, 1.08)
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', frameon=True, fontsize=9)
fig.tight_layout()
fig.savefig('figures/federated_power_curves.pdf', dpi=300, bbox_inches='tight')
fig.savefig('figures/federated_power_curves.png', dpi=220, bbox_inches='tight')
plt.show()
print('Saved figures/federated_power_curves.{pdf,png}')

# === Figure: headline plot at R² = 0.15 (medium effect size) ===
r2_pick = 0.15
f2_pick = f2_for_r2(r2_pick)
fig, ax = plt.subplots(figsize=(9, 5.5))
powers_path = [calc_power(n, K_FEATURES, f2_pick) for n in cohort_sizes]
xs = list(range(len(cohort_sizes)))
ax.plot(xs, powers_path, marker='o', linewidth=2.5, markersize=14, color='#2ca02c')
for x, n, label, p in zip(xs, cohort_sizes, path_labels, powers_path):
    annotation = f'power = {p:.2f}\n(N = {n})' if not np.isnan(p) else f'(N = {n}, model not fittable)'
    ax.annotate(annotation, xy=(x, 0 if np.isnan(p) else p),
                xytext=(0, 18), textcoords='offset points',
                ha='center', fontsize=10)
ax.axhline(y=0.80, color='black', linestyle=':', alpha=0.6, label='80% power')
ax.set_xticks(xs)
ax.set_xticklabels(path_labels, fontsize=10)
ax.set_xlabel('Federation composition (cumulative subjects)')
ax.set_ylabel(f'Statistical power at effect size R² = {r2_pick}')
ax.set_title(f'Federation takes power from below detection to above 80%\n(multiple regression, k = 9 features, α = 0.05, R² = {r2_pick})')
ax.set_ylim(-0.05, 1.10)
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', frameon=True)
fig.tight_layout()
fig.savefig('figures/federated_power_headline.pdf', dpi=300, bbox_inches='tight')
fig.savefig('figures/federated_power_headline.png', dpi=220, bbox_inches='tight')
plt.show()
print('Saved figures/federated_power_headline.{pdf,png}')
```

### Reading the power numbers

At the cumulative cohort sizes produced by the smallest-first federation
path, the table tells one story very clearly:

- **Alone, with N = 10 in SDY569,** there are not even enough subjects
  to fit a multiple regression with 9 features (we need at least
  N = k + 2 = 11). The model is *not estimable* at the single-institution
  scale.
- **Adding SDY1737 (N = 26)** makes the model fittable but power for a
  medium effect (R² = 0.15) is only 17%. We would miss most real
  effects of this size.
- **Adding SDY797 (N = 75)** brings power to 65%. We can detect medium
  effects more often than we miss them, but below the 80% threshold a
  reviewer would ask for.
- **Adding SDY524 (N = 150)** brings power to 96%. We are now decisively

Equivalently, the **minimum detectable effect size** at 80% power

at N = 26 and R² ≈ 0.19 at N = 75, down to R² ≈ 0.10 at N = 150 (a
biologically plausible effect that single-institution studies would
have missed).

**federation lifts the cohort-size ceiling on what effects we can detect, without
any single institution acquiring more subjects.** It is the same
argument that motivates multi-site clinical trials, except expressed
as a computational protocol rather than as a centralised database.

## Principled federated Lasso with ADMM, and the post-Lasso pipeline

The cells above use **median aggregation** of weights or coefficients across
institutions. That works adequately for OLS and Random Forest, but it
**fragments Lasso's feature selection**: each institution's L1 fit picks a
*different* sparse subset of features on its small heterogeneous cohort, and
the median of "feature X selected by institution 1, not by institution 2" is
the wrong feature half-zeroed.

The right algorithm for federated Lasso is **ADMM consensus** (Alternating
Direction Method of Multipliers; Boyd et al. 2011, *Distributed Optimization
and Statistical Learning via the ADM of Multipliers*, F&T Machine Learning
3(1):1-122). Each institution does a *dense* ridge-style local fit; a single
coordinator step per round applies the L1 penalty to a consensus vector. The
sparsity is enforced *globally*, not per-institution, and the federated
solution provably converges to the centralised Lasso optimum for convex
problems.

This section:

1. Implements ADMM-Lasso with an intercept column (not penalised) and
   adaptive ρ (Boyd 2011 §3.4.1) — needed for heterogeneous study sizes.
2. Verifies it converges to the centralised Lasso solution to machine
   precision on Panel A *and* Panel B (the original implementation had
   convergence issues on the heterogeneous Panel B cohort).
3. Uses ADMM-Lasso as a **federated feature selector**, then refits MLR,
   Lasso, and Random Forest on the selected support — each under its
   correct per-method aggregation rule (FedAvg-by-N, ADMM, union of forests).
4. Runs a **sensitivity analysis** excluding SDY1737, which has the most
   distinct clinical profile (older age stratum, longer disease duration,
   smaller cohort) and the worst per-study held-out R² under every method.

The headline empirical finding lives in the cleaned Panel B configuration.

```python
# === ADMM-Lasso v2: intercept column + adaptive rho ===
# Boyd 2011 §3.4.1 consensus form. f_k normalised by 1/(2 N_total) so the
# alpha here matches sklearn.linear_model.Lasso's convention.

import sys
sys.path.insert(0, "src")
import oadr_data as od
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import StratifiedKFold
from scipy.stats import f as fdist, ncf

def soft_threshold(v, lam):
    return np.sign(v) * np.maximum(np.abs(v) - lam, 0.0)

def admm_lasso_v2(X_list, y_list, alpha, rho_init=1.0, max_iter=10000, tol=1e-7,
                  adaptive_rho=True, mu=10.0, tau=2.0):
    """
    Consensus ADMM-Lasso with intercept column.

    Solves (sklearn-Lasso objective):
        (1/(2 N_total)) || y - X beta ||^2 + alpha || beta_non_intercept ||_1

    Parameters
    ----------
    X_list, y_list : lists of (X_k, y_k) per institution. X_k must be
        feature-scaled, but should NOT include an intercept column — we add it.
    alpha : L1 penalty strength (sklearn convention).
    rho_init : initial Lagrangian penalty.
    adaptive_rho : if True, doubles rho when primal residual >> dual and
        halves it when opposite, every 20 iterations.

    Returns
    -------
    intercept, coefs, n_iter
    """
    K = len(X_list)
    Xa_list = [np.column_stack([np.ones(len(X)), X]) for X in X_list]
    p = Xa_list[0].shape[1]
    N = sum(len(y) for y in y_list)
    betas = [np.zeros(p) for _ in range(K)]
    us = [np.zeros(p) for _ in range(K)]
    z = np.zeros(p)
    rho = rho_init
    Xty = [(1.0/N) * (Xa.T @ y) for Xa, y in zip(Xa_list, y_list)]
    XtX = [(1.0/N) * (Xa.T @ Xa) for Xa in Xa_list]
    def compute_A_inv(rho_val):
        return [np.linalg.inv(XtX[k] + rho_val * np.eye(p)) for k in range(K)]
    A_inv = compute_A_inv(rho)
    for t in range(max_iter):
        for k in range(K):
            betas[k] = A_inv[k] @ (Xty[k] + rho * (z - us[k]))
        mean_term = np.mean([betas[k] + us[k] for k in range(K)], axis=0)
        z_new = mean_term.copy()
        z_new[1:] = soft_threshold(mean_term[1:], alpha / (rho * K))
        # intercept (z_new[0]) is NOT thresholded
        for k in range(K):
            us[k] = us[k] + betas[k] - z_new
        primal = np.sqrt(np.mean([np.sum((betas[k] - z_new)**2) for k in range(K)]))
        dual = rho * np.linalg.norm(z_new - z)
        z = z_new
        if primal < tol and dual < tol:
            return z[0], z[1:], t+1
        if adaptive_rho and (t+1) % 20 == 0:
            if primal > mu * dual:
                rho *= tau
                for k in range(K): us[k] /= tau
                A_inv = compute_A_inv(rho)
            elif dual > mu * primal:
                rho /= tau
                for k in range(K): us[k] *= tau
                A_inv = compute_A_inv(rho)
    return z[0], z[1:], max_iter
```

```python
# === Verify ADMM converges to centralised Lasso on both panels ===
# Panel A at alpha=0.05 (the value that selects 0 features in centralised fit),
# Panel B at alpha=0.008 (selects 6 features matching the talk summary).

A_all = od.load_panel_a_all()
B_all = od.load_panel_b_all()

Xa = A_all[od.PANEL_A_FEATURES].values.astype(float)
ya = A_all[od.PANEL_A_TARGET].values.astype(float)
studies_a = A_all["Study"].values
fa = list(od.PANEL_A_FEATURES)

Xb_df, yb_s, fb = od.panel_b_design_matrix(B_all)
Xb = Xb_df.values.astype(float); yb = yb_s.values.astype(float)
studies_b = B_all["Study"].values
fb = list(fb)

def verify_admm(X, y, studies, fnames, alpha, label):
    sc = MinMaxScaler().fit(X)
    m_pool = Lasso(alpha=alpha, max_iter=20000).fit(sc.transform(X), y)
    X_list = [sc.transform(X[studies==s]) for s in sorted(set(studies))]
    y_list = [y[studies==s] for s in sorted(set(studies))]
    intc, coefs, n_iter = admm_lasso_v2(X_list, y_list, alpha=alpha)
    diff = np.max(np.abs(coefs - m_pool.coef_))
    print(f"{label}: pooled non-zero = {(np.abs(m_pool.coef_)>1e-6).sum()}/{len(fnames)}, "
          f"ADMM non-zero = {(np.abs(coefs)>1e-6).sum()}/{len(fnames)}, "
          f"||ADMM - pooled||_inf = {diff:.2e}, n_iter = {n_iter}")
    return coefs, m_pool.coef_

print("ADMM-Lasso convergence vs centralised reference:")
print()
ca_admm, ca_pool = verify_admm(Xa, ya, studies_a, fa, alpha=0.05, label="Panel A (alpha=0.05)")
cb_admm, cb_pool = verify_admm(Xb, yb, studies_b, fb, alpha=0.008, label="Panel B (alpha=0.008)")
print()
print("ADMM recovers the centralised Lasso solution at machine precision on both panels.")
print("Median aggregation does NOT, and selection collapses — see the lasso_aggregation_comparison figure.")
```

### Post-Lasso pipeline — refit all methods on the ADMM-selected support

ADMM-Lasso gives a federation-consistent feature selection. Once the
support is identified, the remaining question is: how do the other methods
predict on that reduced feature set? Each method gets its own correct
aggregation rule:

| Method | Aggregation | Why |
|---|---|---|
| Multiple Linear (OLS on selected support) | **FedAvg-by-N** (weighted mean) | OLS decomposes into per-institution sufficient statistics; weighted mean recovers the pooled OLS exactly |
| Lasso on selected support | **ADMM** (re-run at the same alpha) | Same logic as the selection step; cross-validates the selection |
| Random Forest | **Union of forests** (each test point predicted by averaging over every institution's forest) | RF is an ensemble; no parameters to aggregate |

CV protocol: `StratifiedKFold(n_splits=5)` on `Study`, identical to the
existing federated experiment matrix. SDY1625 has no C-peptide data and is
absent from both panels' supervised work.

```python
# === Run the pipeline across 4 configurations ===
#   Panel A, all 4 studies
#   Panel A, drop SDY1737  (sensitivity analysis)
#   Panel B, all 3 studies
#   Panel B, drop SDY1737  (sensitivity analysis)

def calc_power(n, k, f2, alpha=0.05):
    if n <= k+1 or f2 <= 0: return float("nan")
    F_crit = fdist.ppf(1-alpha, k, n-k-1)
    nc = f2 * n
    return float(1 - ncf.cdf(F_crit, k, n-k-1, nc))

def f2_from_r2(r2):
    return r2/(1-r2) if 0 < r2 < 1 else 0.0

def federated_oof(X, y, studies, method, alpha=None):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    oof = np.full_like(y, np.nan)
    for tr, te in skf.split(X, studies):
        local_models, scalers = {}, {}
        for s in sorted(set(studies[tr])):
            idx = tr[studies[tr]==s]
            if len(idx) < 2: continue
            sc = MinMaxScaler().fit(X[idx])
            Xs = sc.transform(X[idx])
            scalers[s] = sc
            if method == "mlr_fedavg":
                m = LinearRegression().fit(Xs, y[idx])
                local_models[s] = (m.coef_, m.intercept_, len(idx))
            elif method == "rf_union":
                m = RandomForestRegressor(n_estimators=200, min_samples_leaf=2,
                                          n_jobs=1, random_state=42).fit(Xs, y[idx])
                local_models[s] = m
        if method == "mlr_fedavg":
            coefs = np.stack([local_models[s][0] for s in local_models])
            ints = np.array([local_models[s][1] for s in local_models])
            sizes = np.array([local_models[s][2] for s in local_models])
            agg_coef = np.average(coefs, axis=0, weights=sizes)
            agg_int = np.average(ints, weights=sizes)
            for i in te:
                s = studies[i]
                if s not in scalers: continue
                xs = scalers[s].transform(X[i:i+1])[0]
                oof[i] = agg_coef @ xs + agg_int
        elif method == "lasso_admm":
            sc_g = MinMaxScaler().fit(X[tr])
            X_g = [sc_g.transform(X[tr[studies[tr]==s]]) for s in sorted(set(studies[tr]))]
            y_g = [y[tr[studies[tr]==s]] for s in sorted(set(studies[tr]))]
            intc, coef, _ = admm_lasso_v2(X_g, y_g, alpha=alpha or 0.008)
            for i in te:
                xs = sc_g.transform(X[i:i+1])[0]
                oof[i] = coef @ xs + intc
        elif method == "rf_union":
            for i in te:
                preds = []
                for s_k, m in local_models.items():
                    xs = scalers[s_k].transform(X[i:i+1])
                    preds.append(m.predict(xs)[0])
                oof[i] = np.mean(preds)
    return oof

def metrics(y, oof):
    mask = ~np.isnan(oof)
    mse = np.mean((y[mask]-oof[mask])**2)
    rss = np.sum((y[mask]-oof[mask])**2); tss = np.sum((y[mask]-y[mask].mean())**2)
    r2 = 1 - rss/tss if tss>0 else float("nan")
    return mse, r2, int(mask.sum())

ALPHA_PANEL_A = 0.02
ALPHA_PANEL_B = 0.008

configs = [
    ("Panel A, all 4 studies", Xa, ya, studies_a, fa, ALPHA_PANEL_A),
    ("Panel A, drop SDY1737",  Xa[studies_a!="SDY1737"], ya[studies_a!="SDY1737"],
                              studies_a[studies_a!="SDY1737"], fa, ALPHA_PANEL_A),
    ("Panel B, all 3 studies", Xb, yb, studies_b, fb, ALPHA_PANEL_B),
    ("Panel B, drop SDY1737",  Xb[studies_b!="SDY1737"], yb[studies_b!="SDY1737"],
                              studies_b[studies_b!="SDY1737"], fb, ALPHA_PANEL_B),
]

post_lasso_rows = []
for name, X, y, st, fnames, alpha in configs:
    print(f"\n--- {name}  N={len(X)}, studies={dict(pd.Series(st).value_counts())} ---")
    sc_g = MinMaxScaler().fit(X)
    X_list = [sc_g.transform(X[st==s]) for s in sorted(set(st))]
    y_list = [y[st==s] for s in sorted(set(st))]
    intc, coef, n_iter = admm_lasso_v2(X_list, y_list, alpha=alpha)
    sel = np.abs(coef) > 1e-6
    sel_names = [f for f, k in zip(fnames, sel) if k]
    print(f"  ADMM-Lasso selected {sel.sum()}/{len(fnames)} features after {n_iter} iter: {sel_names}")
    if sel.sum() == 0:
        continue
    X_sel = X[:, sel]
    for method, label in [("mlr_fedavg","MLR FedAvg-by-N"),
                          ("lasso_admm","Lasso ADMM"),
                          ("rf_union",  "RF union-of-forests")]:
        a = alpha if method == "lasso_admm" else None
        oof = federated_oof(X_sel, y, st, method, alpha=a)
        mse, r2, n = metrics(y, oof)
        power = calc_power(n, int(sel.sum()), f2_from_r2(r2))
        print(f"    {label:<22s}  MSE={mse:.3f}  R²={r2:+.3f}  power={power:.3f}")
        post_lasso_rows.append({"config":name, "method":label,
                                "n_selected":int(sel.sum()), "MSE":mse, "R2":r2,
                                "achieved_power":power, "N":n,
                                "selected":", ".join(sel_names)})

post_lasso_df = pd.DataFrame(post_lasso_rows)
post_lasso_df.to_csv("results/admm_post_lasso_pipeline.csv", index=False)
print("\nSaved: results/admm_post_lasso_pipeline.csv")
```

### Lasso aggregation rule comparison: Median vs FedAvg-by-N vs ADMM vs Pooled

Four aggregation rules evaluated on Panel A (9 features, N=150, 4 institutions,
α=0.05, 5-fold stratified-by-Study CV):

| Rule | Description | Deployable? |
|---|---|---|
| Pooled Lasso | Centralised fit on all data combined | No — requires data sharing |
| Median federation | Per-institution local Lasso, median of coefficients | Yes |
| FedAvg-by-N | Per-institution local Lasso, weighted mean of coefficients (weights ∝ training-set size) | Yes |
| ADMM consensus | Coordinated optimisation per Boyd 2011 §8.2 | Yes |

The pooled column is shown as a reference point only — it is the centralised
solution the federated rules are trying to approximate.

```python
# === Lasso aggregation comparison: 5-fold CV at alpha=0.05 ===

ALPHA = 0.05
RNG = 42

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RNG)
rows = []
for fold_idx, (tr, te) in enumerate(skf.split(Xa, studies_a)):
    # Pooled
    sc_p = MinMaxScaler().fit(Xa[tr])
    m_p = Lasso(alpha=ALPHA, max_iter=20000).fit(sc_p.transform(Xa[tr]), ya[tr])
    pred_pooled = m_p.predict(sc_p.transform(Xa[te]))

    # Per-study local fits for median + FedAvg
    local_coefs, local_intercepts, local_sizes, scalers = {}, {}, {}, {}
    X_list, y_list, study_order = [], [], []
    for s in sorted(set(studies_a[tr])):
        idx = tr[studies_a[tr] == s]
        if len(idx) < 2: continue
        sc = MinMaxScaler().fit(Xa[idx])
        Xs = sc.transform(Xa[idx])
        m_s = Lasso(alpha=ALPHA, max_iter=20000).fit(Xs, ya[idx])
        local_coefs[s] = m_s.coef_; local_intercepts[s] = m_s.intercept_
        local_sizes[s] = len(idx); scalers[s] = sc
        X_list.append(Xs); y_list.append(ya[idx]); study_order.append(s)

    coef_stack = np.stack([local_coefs[s] for s in study_order])
    int_stack = np.array([local_intercepts[s] for s in study_order])
    sizes = np.array([local_sizes[s] for s in study_order])
    med_coef = np.median(coef_stack, axis=0); med_int = np.median(int_stack)
    avg_coef = np.average(coef_stack, axis=0, weights=sizes); avg_int = np.average(int_stack, weights=sizes)

    # ADMM consensus, fit per fold with the global scaler
    sc_g = MinMaxScaler().fit(Xa[tr])
    X_g = [sc_g.transform(Xa[tr[studies_a[tr]==s]]) for s in study_order]
    y_g = [ya[tr[studies_a[tr]==s]] for s in study_order]
    admm_intc, admm_coef, admm_iter = admm_lasso_v2(X_g, y_g, alpha=ALPHA)

    # Predict each test subject with its own study's scaler (except for ADMM which uses sc_g)
    pred_med = np.zeros(len(te)); pred_avg = np.zeros(len(te)); pred_admm = np.zeros(len(te))
    for j, i in enumerate(te):
        s = studies_a[i]
        sc = scalers.get(s)
        if sc is None:
            pred_med[j] = pred_avg[j] = pred_admm[j] = np.nan; continue
        xs = sc.transform(Xa[i:i+1])[0]
        pred_med[j] = med_coef @ xs + med_int
        pred_avg[j] = avg_coef @ xs + avg_int
        xs_g = sc_g.transform(Xa[i:i+1])[0]
        pred_admm[j] = admm_coef @ xs_g + admm_intc

    mask = ~np.isnan(pred_med)
    rows.append({
        "fold": fold_idx+1,
        "Lasso_pooled": np.mean((ya[te] - pred_pooled)**2),
        "Lasso_FedMedian": np.mean((ya[te][mask] - pred_med[mask])**2),
        "Lasso_FedAvg_byN": np.mean((ya[te][mask] - pred_avg[mask])**2),
        "Lasso_ADMM": np.mean((ya[te][mask] - pred_admm[mask])**2),
        "ADMM_n_iter": admm_iter,
    })

agg_df = pd.DataFrame(rows)
agg_df.to_csv("results/lasso_aggregation_comparison.csv", index=False)
print("Mean MSE per aggregation rule (5-fold CV, Panel A, alpha=0.05):")
print(agg_df.drop(columns=["fold", "ADMM_n_iter"]).mean().round(3))

# Selection counts on the full data
sc_full = MinMaxScaler().fit(Xa)
m_full = Lasso(alpha=ALPHA, max_iter=20000).fit(sc_full.transform(Xa), ya)
local_full = {}
for s in sorted(set(studies_a)):
    idx = (studies_a == s).nonzero()[0]
    sc = MinMaxScaler().fit(Xa[idx])
    m_s = Lasso(alpha=ALPHA, max_iter=20000).fit(sc.transform(Xa[idx]), ya[idx])
    local_full[s] = m_s.coef_
coefs_full = np.stack(list(local_full.values()))
sizes_full = np.array([(studies_a == s).sum() for s in local_full.keys()])
med_full = np.median(coefs_full, axis=0)
avg_full = np.average(coefs_full, axis=0, weights=sizes_full)
Xg = [MinMaxScaler().fit(Xa).transform(Xa[studies_a==s]) for s in sorted(set(studies_a))]
yg = [ya[studies_a==s] for s in sorted(set(studies_a))]
_, admm_full, _ = admm_lasso_v2(Xg, yg, alpha=ALPHA)
selected = {
    "Lasso_pooled":    int((np.abs(m_full.coef_) > 1e-6).sum()),
    "Lasso_FedMedian": int((np.abs(med_full) > 1e-6).sum()),
    "Lasso_FedAvg_byN":int((np.abs(avg_full) > 1e-6).sum()),
    "Lasso_ADMM":      int((np.abs(admm_full) > 1e-6).sum()),
}
print()
print("Number of features kept (out of 9):")
for k, v in selected.items():
    print(f"  {k:<20s}: {v}/9")
```

### Statistical power: theoretical curves with empirical reference lines

The power calculation in cell 58 produces curves at assumed effect sizes —
power as a function of cohort size IF a true effect of the assumed magnitude
exists in the population. The figures below add empirical reference lines:
the held-out R² we actually observed on Panel A (≈ 0, no detectable signal
on autoantibodies alone) and Panel B (0.09 with Lasso, small positive
effect). This makes clear that federation lifts the design's *capability*
to detect effects (the ceiling), while the empirical signal in the data
sets the floor.

## Per-method scatter figures: how a small institution improves under federation

Each of the next nine figures has the same structure: a two-panel scatter of
predicted versus observed log(C-peptide AUC).

- **Left panel** — the smallest cohort (SDY569, N=10) trained on its own. This
  is what one institution can do alone, with no federation.
- **Right panel** — all participating institutions trained together under
  federation. Each point is a held-out subject from 5-fold stratified-by-Study
  cross-validation, colour-coded by institution.

The point of these figures is to show the *visual* effect of going from a tiny
solo cohort to a federated cohort: how the predictions tighten around the
y = x line, and where each institution lands in the federated result.

Below we define two small helpers used by every tabular-method figure
(MLR, Lasso, Random Forest). The CNN figures (plain CNN and AE+CNN) need
their own training loops and are written out individually.

### Helper 1 — `federated_scatter_oof`

Given a feature matrix `X`, a target `y`, an array of per-subject study
labels, and a method name, this function returns two arrays of out-of-fold
predictions:

- `oof_solo`: predictions for SDY569 subjects when the model is trained only
  on SDY569's own training-fold subjects (no other institution contributes).
- `oof_fed`: predictions for *every* subject when the model is trained on
  the union of training-fold subjects across all institutions, federated
  by the method's proper aggregation rule (FedAvg-by-N for MLR, ADMM for
  Lasso, union of forests for Random Forest).

Both arrays use the same 5 cross-validation folds, so the SDY569 dots in
the left and right panels of the resulting figure are the *same subjects
predicted under two protocols* — the apples-to-apples comparison.

```python
def federated_scatter_oof(X, y, study_labels, method, alpha=None, rng_seed=42):
    """
    Out-of-fold predictions for the SDY569-alone-vs-federated scatter figures.

    Parameters
    ----------
    X : array, shape (n_subjects, n_features). Feature matrix, raw values.
    y : array, shape (n_subjects,). Target (log C-peptide AUC).
    study_labels : array of strings, shape (n_subjects,). Per-subject institution.
    method : str. One of {"mlr_fedavg", "lasso_admm", "rf_union"}.
    alpha : float or None. L1 strength for Lasso ADMM (ignored otherwise).

    Returns
    -------
    oof_solo, oof_fed : two arrays of shape (n_subjects,) with NaN at
        positions that are not in the relevant test set.
    """
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=rng_seed)
    oof_solo = np.full_like(y, np.nan, dtype=float)
    oof_fed  = np.full_like(y, np.nan, dtype=float)

    for tr, te in skf.split(X, study_labels):
        # ---- SDY569 alone, this fold ----
        tr569 = tr[study_labels[tr] == "SDY569"]
        te569 = te[study_labels[te] == "SDY569"]
        if len(tr569) >= 2 and len(te569) >= 1:
            sc = MinMaxScaler().fit(X[tr569])
            Xs = sc.transform(X[tr569])
            if method == "mlr_fedavg":
                m = LinearRegression().fit(Xs, y[tr569])
            elif method == "lasso_admm":
                m = Lasso(alpha=alpha or 0.008, max_iter=20000).fit(Xs, y[tr569])
            elif method == "rf_union":
                m = RandomForestRegressor(n_estimators=200, min_samples_leaf=2,
                                          n_jobs=1, random_state=42).fit(Xs, y[tr569])
            oof_solo[te569] = m.predict(sc.transform(X[te569]))

        # ---- All institutions federated, this fold ----
        local_models, scalers = {}, {}
        X_list, y_list, sord = [], [], []
        for s in sorted(set(study_labels[tr])):
            idx = tr[study_labels[tr] == s]
            if len(idx) < 2: continue
            sc = MinMaxScaler().fit(X[idx])
            Xs = sc.transform(X[idx])
            scalers[s] = sc
            if method == "mlr_fedavg":
                m = LinearRegression().fit(Xs, y[idx])
                local_models[s] = (m.coef_, m.intercept_, len(idx))
            elif method == "lasso_admm":
                X_list.append(Xs); y_list.append(y[idx]); sord.append(s)
            elif method == "rf_union":
                m = RandomForestRegressor(n_estimators=200, min_samples_leaf=2,
                                          n_jobs=1, random_state=42).fit(Xs, y[idx])
                local_models[s] = m

        if method == "mlr_fedavg":
            coefs = np.stack([local_models[s][0] for s in local_models])
            ints  = np.array([local_models[s][1] for s in local_models])
            sizes = np.array([local_models[s][2] for s in local_models])
            agg_c = np.average(coefs, axis=0, weights=sizes)
            agg_i = np.average(ints, weights=sizes)
            for i in te:
                if study_labels[i] in scalers:
                    xs = scalers[study_labels[i]].transform(X[i:i+1])[0]
                    oof_fed[i] = agg_c @ xs + agg_i
        elif method == "lasso_admm":
            sc_g = MinMaxScaler().fit(X[tr])
            X_g = [sc_g.transform(X[tr[study_labels[tr]==s]]) for s in sord]
            y_g = [y[tr[study_labels[tr]==s]] for s in sord]
            intc, coef, _ = admm_lasso_v2(X_g, y_g, alpha=alpha or 0.008)
            for i in te:
                xs = sc_g.transform(X[i:i+1])[0]
                oof_fed[i] = coef @ xs + intc
        elif method == "rf_union":
            for i in te:
                preds = []
                for s_k, m in local_models.items():
                    xs = scalers[s_k].transform(X[i:i+1])
                    preds.append(m.predict(xs)[0])
                if preds:
                    oof_fed[i] = np.mean(preds)
    return oof_solo, oof_fed
```

### Helper 2 — `plot_scatter_pair`

Draws the two-panel figure given the out-of-fold predictions. Left panel
shows SDY569 alone; right panel shows all institutions with one colour per
institution. Each panel reports MSE and R² in its title.

```python
def plot_scatter_pair(y, study_labels, oof_solo, oof_fed,
                      suptitle, out_path,
                      study_colors=None):
    """
    Two-panel scatter: left = SDY569 alone, right = all institutions federated.

    Saves to `out_path` (PNG; also writes a matching .pdf).
    """
    if study_colors is None:
        study_colors = {"SDY524":"#1f77b4", "SDY569":"#d62728",
                        "SDY797":"#2ca02c", "SDY1737":"#9467bd"}

    def mse_r2(yt, yp):
        m = ~np.isnan(yp)
        if m.sum() < 2: return float("nan"), float("nan")
        mse = float(np.mean((yt[m]-yp[m])**2))
        rss = float(np.sum((yt[m]-yp[m])**2))
        tss = float(np.sum((yt[m]-yt[m].mean())**2))
        r2 = 1 - rss/tss if tss > 0 else float("nan")
        return mse, r2

    is569 = (study_labels == "SDY569")
    mse_solo, r2_solo = mse_r2(y[is569], oof_solo[is569])
    mse_fed,  r2_fed  = mse_r2(y, oof_fed)

    fig, axes = plt.subplots(1, 2, figsize=(11.5, 5), constrained_layout=True)

    # Left: SDY569 alone
    ax = axes[0]
    mask = is569 & ~np.isnan(oof_solo)
    ax.scatter(y[mask], oof_solo[mask], c=study_colors["SDY569"], s=65,
               edgecolor="white", linewidth=0.6, label=f"SDY569 (N={mask.sum()})")
    if mask.sum() > 0:
        lo = min(y[mask].min(), oof_solo[mask].min())
        hi = max(y[mask].max(), oof_solo[mask].max())
        ax.plot([lo,hi],[lo,hi], "k--", alpha=0.4, label="y = x")
    ax.set_xlabel("Observed log(C-peptide AUC)")
    ax.set_ylabel("Predicted log(C-peptide AUC)")
    ax.set_title(f"SDY569 alone  (N = {mask.sum()})\n"
                 f"MSE = {mse_solo:.3f}    R² = {r2_solo:+.3f}", fontweight="bold")
    ax.legend(loc="lower right", fontsize=9); ax.grid(alpha=0.25)

    # Right: all institutions federated
    ax = axes[1]
    for s in sorted(set(study_labels)):
        mask = (study_labels == s) & ~np.isnan(oof_fed)
        if mask.sum() == 0: continue
        ax.scatter(y[mask], oof_fed[mask], c=study_colors.get(s, "grey"),
                   s=65, edgecolor="white", linewidth=0.6, alpha=0.85,
                   label=f"{s} (N={mask.sum()})")
    all_mask = ~np.isnan(oof_fed)
    if all_mask.sum() > 0:
        lo = min(y[all_mask].min(), oof_fed[all_mask].min())
        hi = max(y[all_mask].max(), oof_fed[all_mask].max())
        ax.plot([lo,hi],[lo,hi], "k--", alpha=0.4, label="y = x")
    ax.set_xlabel("Observed log(C-peptide AUC)")
    ax.set_ylabel("Predicted log(C-peptide AUC)")
    ax.set_title(f"All institutions federated  (N = {all_mask.sum()})\n"
                 f"MSE = {mse_fed:.3f}    R² = {r2_fed:+.3f}", fontweight="bold")
    ax.legend(loc="lower right", fontsize=9); ax.grid(alpha=0.25)

    fig.suptitle(suptitle, fontsize=13, fontweight="bold")
    fig.savefig(out_path, dpi=220)
    fig.savefig(out_path.replace(".png", ".pdf"))
    plt.show()
    return mse_solo, r2_solo, mse_fed, r2_fed
```

### Figure 4 — Lasso on Panel B with naive median aggregation (diagnostic)

Lasso on Panel B at α = 0.05, but here we deliberately use the **naive
median aggregation** rule rather than ADMM. This is the figure that flagged
the "cloud" of poorly-predicted points and led us to investigate the SDY1737
cohort. Keep this for reference; the recommended ADMM version is the next
figure.

What to look for: the right panel has a wide horizontal band of predictions
clustered between roughly −0.7 and −1.2, regardless of the observed value.
That band is the signature of median federation fragmenting Lasso's feature
selection.

```python
# Median-aggregation Lasso, per-fold. This is the diagnostic figure; the
# recommended aggregation rule (ADMM) is the next figure.
def federated_scatter_lasso_median(X, y, study_labels, alpha=0.05, seed=42):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    oof_solo = np.full_like(y, np.nan, dtype=float)
    oof_fed  = np.full_like(y, np.nan, dtype=float)
    for tr, te in skf.split(X, study_labels):
        tr569 = tr[study_labels[tr]=="SDY569"]; te569 = te[study_labels[te]=="SDY569"]
        if len(tr569) >= 2 and len(te569) >= 1:
            sc = MinMaxScaler().fit(X[tr569])
            m = Lasso(alpha=alpha, max_iter=20000).fit(sc.transform(X[tr569]), y[tr569])
            oof_solo[te569] = m.predict(sc.transform(X[te569]))
        local_coefs, local_ints, scalers = {}, {}, {}
        for s in sorted(set(study_labels[tr])):
            idx = tr[study_labels[tr]==s]
            if len(idx) < 2: continue
            sc = MinMaxScaler().fit(X[idx])
            m = Lasso(alpha=alpha, max_iter=20000).fit(sc.transform(X[idx]), y[idx])
            local_coefs[s] = m.coef_; local_ints[s] = m.intercept_; scalers[s] = sc
        coefs = np.stack(list(local_coefs.values()))
        med_coef = np.median(coefs, axis=0)
        med_int = np.median(list(local_ints.values()))
        for i in te:
            if study_labels[i] in scalers:
                xs = scalers[study_labels[i]].transform(X[i:i+1])[0]
                oof_fed[i] = med_coef @ xs + med_int
    return oof_solo, oof_fed

oof_solo_lb_med, oof_fed_lb_med = federated_scatter_lasso_median(Xb, yb, studies_b, alpha=0.05)
plot_scatter_pair(yb, studies_b, oof_solo_lb_med, oof_fed_lb_med,
                  suptitle="Federation with Lasso — Panel B (19 features), MEDIAN aggregation (diagnostic)",
                  out_path="figures/scatter_lasso_panelB.png")
```

### Figure 8 — Plain CNN on Panel A (no autoencoder)

The 9 Panel A features are reshaped into a 3×3 spatial grid for Conv2D. The
architecture is Conv2D(8, 3×3, relu) → Flatten → Dense(4, relu) → Dense(16, relu)
→ Dense(1). Federation rule is median aggregation of weights across
institutions.

**Heads-up**: this cell trains five small CNNs per fold (one per institution)
and may take roughly two to three minutes.

What to look for: median-of-weights aggregation of independently-initialised
neural networks tends to settle into a bad parameter regime, especially
when each institution lands in a different basin of the loss landscape. We
expect federation to *hurt* the plain CNN relative to solo training.

```python
# Plain CNN scatter — Panel A only (3x3 reshape requires 9 features)
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
import tensorflow as tf

def to_3x3(X9):
    return X9.reshape(-1, 3, 3, 1).astype(np.float32)

def build_plain_cnn(seed):
    tf.keras.utils.set_random_seed(seed)
    m = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(3,3,1)),
        tf.keras.layers.Conv2D(8, 3, activation="relu", padding="same"),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(4, activation="relu"),
        tf.keras.layers.Dense(16, activation="relu"),
        tf.keras.layers.Dense(1),
    ])
    m.compile(optimizer="adam", loss="mse")
    return m

def median_weights(ws):
    return [np.median(np.stack(w, axis=0), axis=0) for w in zip(*ws)]

RNG = 42
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RNG)
oof_solo_cnn = np.full_like(ya, np.nan, dtype=float)
oof_fed_cnn  = np.full_like(ya, np.nan, dtype=float)
for fold_idx, (tr, te) in enumerate(skf.split(Xa, studies_a)):
    seed = RNG + fold_idx
    # Solo SDY569
    tr569 = tr[studies_a[tr]=="SDY569"]; te569 = te[studies_a[te]=="SDY569"]
    if len(tr569) >= 2 and len(te569) >= 1:
        sc = MinMaxScaler().fit(Xa[tr569])
        m = build_plain_cnn(seed)
        m.fit(to_3x3(sc.transform(Xa[tr569])), ya[tr569], epochs=100, batch_size=8, verbose=0)
        oof_solo_cnn[te569] = m.predict(to_3x3(sc.transform(Xa[te569])), verbose=0).flatten()
    # Federated: per-study local fits, median aggregate
    weights, scalers = [], {}
    for s in sorted(set(studies_a[tr])):
        idx = tr[studies_a[tr]==s]
        if len(idx) < 2: continue
        sc = MinMaxScaler().fit(Xa[idx])
        m = build_plain_cnn(seed)
        m.fit(to_3x3(sc.transform(Xa[idx])), ya[idx], epochs=100, batch_size=8, verbose=0)
        weights.append(m.get_weights()); scalers[s] = sc
    fed = build_plain_cnn(seed); fed.set_weights(median_weights(weights))
    for i in te:
        s = studies_a[i]
        if s in scalers:
            xs = to_3x3(scalers[s].transform(Xa[i:i+1]))
            oof_fed_cnn[i] = fed.predict(xs, verbose=0).flatten()[0]

plot_scatter_pair(ya, studies_a, oof_solo_cnn, oof_fed_cnn,
                  suptitle="Federation with plain CNN (no autoencoder) — Panel A (9 features)",
                  out_path="figures/scatter_plain_cnn_panelA.png")
```

### Figure 9 — Autoencoder + CNN head on Panel A (the rescue)

Same 3×3 input as Figure 8, but the supervised head sits on top of a
*frozen* federated autoencoder. The autoencoder is trained per institution
on reconstruction (unsupervised), then the encoders are median-aggregated
into a federated encoder. Each institution then trains a small supervised
head on top of the *federated* encoder; the heads are aggregated by median
too.

**Heads-up**: this cell trains both autoencoders and heads per fold, so it
takes longer than the plain CNN — roughly four to six minutes total.

What to look for: the autoencoder pretraining gives every institution a
shared latent space, which makes weight-level aggregation reasonable for
the supervised head on top. Federation should *help* (not hurt) the AE+CNN
relative to the plain CNN.

```python
def build_ae(seed):
    tf.keras.utils.set_random_seed(seed)
    inp = tf.keras.Input(shape=(3,3,1))
    x = tf.keras.layers.Conv2D(8, 3, activation="relu", padding="same")(inp)
    x = tf.keras.layers.Flatten()(x)
    latent = tf.keras.layers.Dense(4, activation="relu", name="latent")(x)
    x2 = tf.keras.layers.Dense(9, activation="relu")(latent)
    x2 = tf.keras.layers.Reshape((3,3,1))(x2)
    out = tf.keras.layers.Conv2D(1, 3, activation="linear", padding="same")(x2)
    ae = tf.keras.Model(inp, out); ae.compile(optimizer="adam", loss="mse")
    encoder = tf.keras.Model(inp, latent)
    return ae, encoder

def build_head_on_frozen_encoder(enc_weights, seed):
    tf.keras.utils.set_random_seed(seed)
    _, encoder = build_ae(seed)
    n_enc = len(encoder.get_weights())
    encoder.set_weights(enc_weights[:n_enc])
    for layer in encoder.layers:
        layer.trainable = False
    h = tf.keras.layers.Dense(16, activation="relu")(encoder.output)
    out = tf.keras.layers.Dense(1)(h)
    head = tf.keras.Model(encoder.input, out)
    head.compile(optimizer="adam", loss="mse")
    return head, encoder

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RNG)
oof_solo_ae = np.full_like(ya, np.nan, dtype=float)
oof_fed_ae  = np.full_like(ya, np.nan, dtype=float)
for fold_idx, (tr, te) in enumerate(skf.split(Xa, studies_a)):
    seed = RNG + fold_idx
    # Solo SDY569: train AE, freeze encoder, train head, predict
    tr569 = tr[studies_a[tr]=="SDY569"]; te569 = te[studies_a[te]=="SDY569"]
    if len(tr569) >= 2 and len(te569) >= 1:
        sc = MinMaxScaler().fit(Xa[tr569])
        Xt = to_3x3(sc.transform(Xa[tr569]))
        ae, enc = build_ae(seed)
        ae.fit(Xt, Xt, epochs=50, batch_size=8, verbose=0)
        head, _ = build_head_on_frozen_encoder(enc.get_weights(), seed)
        head.fit(Xt, ya[tr569], epochs=100, batch_size=8, verbose=0)
        oof_solo_ae[te569] = head.predict(to_3x3(sc.transform(Xa[te569])), verbose=0).flatten()

    # Federated: per-study AE then per-study head on federated encoder
    enc_weights_list, scalers = [], {}
    for s in sorted(set(studies_a[tr])):
        idx = tr[studies_a[tr]==s]
        if len(idx) < 2: continue
        sc = MinMaxScaler().fit(Xa[idx])
        Xt = to_3x3(sc.transform(Xa[idx]))
        ae_s, enc_s = build_ae(seed)
        ae_s.fit(Xt, Xt, epochs=50, batch_size=8, verbose=0)
        enc_weights_list.append(enc_s.get_weights()); scalers[s] = sc
    fed_enc_w = median_weights(enc_weights_list)

    head_weights_list = []
    for s in sorted(set(studies_a[tr])):
        idx = tr[studies_a[tr]==s]
        if len(idx) < 2: continue
        Xt = to_3x3(scalers[s].transform(Xa[idx]))
        head_s, _ = build_head_on_frozen_encoder(fed_enc_w, seed)
        head_s.fit(Xt, ya[idx], epochs=100, batch_size=8, verbose=0)
        head_weights_list.append(head_s.get_weights())
    fed_head_w = median_weights(head_weights_list)
    fed_head, _ = build_head_on_frozen_encoder(fed_enc_w, seed)
    fed_head.set_weights(fed_head_w)
    for i in te:
        s = studies_a[i]
        if s in scalers:
            xs = to_3x3(scalers[s].transform(Xa[i:i+1]))
            oof_fed_ae[i] = fed_head.predict(xs, verbose=0).flatten()[0]

plot_scatter_pair(ya, studies_a, oof_solo_ae, oof_fed_ae,
                  suptitle="Federation with AE + CNN head — Panel A (9 features)",
                  out_path="figures/scatter_ae_cnn_panelA.png")
```

### Figure 12 — Held-out R² on Panel A, all five talk methods

Bar chart of held-out R² across the five methods used in the talk:
multiple linear regression, Lasso, Random Forest, plain CNN, autoencoder +
CNN. Computed under stratified-by-Study 5-fold CV on Panel A.

What to look for: every R² is at-or-below zero. The autoantibody panel
alone does not predict residual β-cell function better than the cohort
mean. This is the empirical motivation for Panel B.

```python
# We have R² for the three tabular methods from earlier in this notebook
# (they were computed inside the post-Lasso pipeline). For the two CNN
# methods we use the held-out R² we just generated in Figures 8 and 9.

def r2_of(yt, yp):
    m = ~np.isnan(yp)
    if m.sum() < 2: return float("nan")
    rss = np.sum((yt[m]-yp[m])**2); tss = np.sum((yt[m]-yt[m].mean())**2)
    return 1 - rss/tss if tss > 0 else float("nan")

panelA_r2 = {
    "Multiple linear": r2_of(ya, oof_fed_a),
    "Lasso (ADMM)":    r2_of(ya, oof_fed_la),
    "Random Forest":   r2_of(ya, oof_fed_ra),
    "Plain CNN":       r2_of(ya, oof_fed_cnn),
    "AE + CNN":        r2_of(ya, oof_fed_ae),
}

fig, ax = plt.subplots(figsize=(9, 5), constrained_layout=True)
methods_x = list(panelA_r2.keys())
vals = [panelA_r2[m] for m in methods_x]
colors_b = ["#1f77b4", "#d62728", "#2ca02c", "#ff7f0e", "#9467bd"]
bars = ax.bar(methods_x, vals, color=colors_b, alpha=0.85, edgecolor="white", linewidth=1.5)
ax.axhline(0, color="black", linewidth=0.8)
ax.set_ylabel("Held-out R² on log(C-peptide AUC)")
ax.set_title("Panel A held-out R² across the five talk methods\nAll values at or below zero — no signal in autoantibodies alone", fontweight="bold")
for b, v in zip(bars, vals):
    if not np.isnan(v):
        yoff = 0.005 if v >= 0 else -0.015
        ax.text(b.get_x()+b.get_width()/2, v+yoff, f"{v:+.3f}",
                ha="center", fontsize=10, fontweight="bold")
ax.grid(alpha=0.3, axis="y")
fig.savefig("figures/panelA_R2_bars.png", dpi=220)
fig.savefig("figures/panelA_R2_bars.pdf")
plt.show()
```
